```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sales_data.csv',encoding='ISO-8859-1')
df
```

```python
df.info()
```

```python
df[df['Amount'].isnull()]
```

```python
df=df.drop(['Status','unnamed1'],axis=1)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
d={'Marital_Status':'Marital'}
df.rename(columns=d,inplace=True)
df
```

```python
df['Marital']=df['Marital'].astype(str)
df.info()
```

```python
d1={'0':'Single','1':'Married'}
df['Marital']=df['Marital'].map(d1)
df
```

```python
d2={'F':'Female','M':'Male'}
df['Gender']=df['Gender'].map(d2)
df
```

```python
sns.countplot(x='Gender',data=df,)
```

```python
plt.figure(figsize=(15,15),facecolor='lightgreen')
plt.gcf().suptitle('Sales Analysis',fontsize=30,color='black',y=1.0)
plt.subplot(3,3,1)
a=sns.countplot(x='Gender',data=df,color='red')
plt.title('Gender count',color='black',fontsize=15)
for i in a.containers:
  a.bar_label(i)

plt.gca().set_facecolor('y')


plt.subplot(3,3,2)
gc=df['Gender'].value_counts()
plt.pie(gc,labels=gc.index,autopct='%1.2f%%',colors=['orange','red','cyan','lightyellow','purple','brown','lightgreen'],shadow=True,textprops={'fontsize':10,'color':'black'},pctdistance=0.7)
plt.title('Gender percentage',color='black',fontsize=15)
plt.gca().set_facecolor('y')



plt.subplot(3,3,3)
sum_gender=df.groupby('Gender',as_index=False)['Amount'].sum()
sum_gender
a=sns.barplot(x='Gender',y='Amount',data=sum_gender,edgecolor='black',hue='Gender',palette='Set1')
plt.title('Gender wise amount',color='black',fontsize=15)
for i in a.containers:
  a.bar_label(i,fmt='%d')

plt.gca().set_facecolor('y')
plt.tight_layout()



plt.subplot(3,3,4)
sum_age=df.groupby('Age Group',as_index=False)['Amount'].sum()
sum_age
a=sns.barplot(x='Age Group',y='Amount',data=sum_age,edgecolor='black',hue='Age Group',palette='Set2')
plt.title('Age group wise amount',color='black',fontsize=15)
for i in a.containers:
  a.bar_label(i,fmt='%d')

plt.gca().set_facecolor('y')
plt.tight_layout()



plt.subplot(3,3,5)
sum_state=df.groupby('State',as_index=False)['Amount'].sum()
sum_state
a=sns.barplot(x='State',y='Amount',data=sum_state,edgecolor='black',hue='State',palette='Set3')
plt.title('State wise amount',color='black',fontsize=15)
for i in a.containers:
  a.bar_label(i,fmt='%d')

plt.xticks(rotation=90)
plt.gca().set_facecolor('y')
plt.tight_layout()


plt.subplot(3,3,6)
sum_occupation =df.groupby('Occupation',as_index=False)['Amount'].sum()
b=sns.barplot(x='Occupation',y='Amount',data=sum_occupation,edgecolor='black',hue='Occupation',palette='Reds')
plt.title('Occupation wise amount',color='black',fontsize=15)
for i in b.containers:
    b.bar_label(i,fmt='%d')

plt.xticks(rotation=90)
plt.gca().set_facecolor('y')
plt.tight_layout()


plt.subplot(3,3,7)
sum_Product_category=df.groupby('Product_Category',as_index=False)['Amount'].sum()
c=sns.barplot(x='Product_Category',y='Amount',data=sum_Product_category,edgecolor='black',hue='Product_Category',palette='coolwarm')
plt.title('Product category wise amount',color='black',fontsize=15)
for i in c.containers:
    c.bar_label(i,fmt='%d')

plt.xticks(rotation=90)
plt.gca().set_facecolor('y')
plt.tight_layout()


plt.subplot(3,3,8)
sum_Marital=df.groupby('Marital',as_index=False)['Amount'].sum()
d=sns.barplot(x='Marital',y='Amount',data=sum_Marital,edgecolor='black',hue='Marital',palette='Reds')
plt.title('Marital status wise amount',color='black',fontsize=15)
for i in d.containers:
    d.bar_label(i,fmt='%d')

plt.xticks(rotation=90)
plt.gca().set_facecolor('y')
plt.tight_layout()



plt.subplot(3,3,9)
f=df['Age Group'].value_counts()
plt.pie(f,labels=f.index,autopct='%0.f%%',colors=['orange','red','cyan','lightyellow','purple','brown','lightgreen'],shadow=True,textprops={'fontsize':10,'color':'black'},pctdistance=0.7)
plt.title('Age group percentage',color='black',fontsize=15)

plt.gca().set_facecolor('y')
plt.tight_layout()
plt.show()

```

Sales Analysis Conclusion:



1.From the first plot, we can see that the number of female customers is higher compared to male customers.

2.The second plot shows that the percentage of female customers is greater than that of male customers.

3.In the third plot, the total purchase amount by females is higher than that of males.

4.The fourth plot indicates that customers in the age group 26 to 35 spend more compared to other age groups.

5.From the fifth plot, we observe that Uttar Pradesh records the highest spending compared to other states.

6.The sixth plot highlights that the IT sector shows the highest spending among different occupations.

7.In the seventh plot, we can see that food products contribute the highest sales compared to other product categories.

8.The eighth plot shows that single customers spend more compared to married customers.

9.Finally, the ninth plot confirms again that the age group 26 to 35 is the most dominant in terms of spending compared to other age groups.
